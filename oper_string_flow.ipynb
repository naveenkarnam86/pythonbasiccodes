Python has a built-in module called random that can be used to make random numbers:

random.randint(a, b): Returns a random integer N such that a <= N <= b. Both a and b are inclusive.    

random.uniform(a, b): Returns a random floating-point number N such that a <= N <= b or b <= N <= a.

```python
import random
print(random.randrange(1,10)) # Generates a random integer between 1 and 10
print(random.uniform(1.5, 5.5)) # Generates a random float between 1.5 and 5.5
```

random.choice(sequence): Returns a randomly selected element from a non-empty sequence (e.g., a list, tuple, or string).

```python
fruits = ["apple", "banana", "orange"]
print(random.choice(fruits))
```

random.shuffle(list): Shuffles the elements of a list in place. This function modifies the original list and returns None.

```python
cards = ["Ace", "King", "Queen"]
random.shuffle(cards)
print(cards)
```

random.seed(a=None): Initializes the random number generator. Providing the same seed value will result in the same sequence of "random" numbers being generated, which is useful for reproducibility in testing or simulations.

```python
random.seed(62) # Set a specific seed
print(random.random())
random.seed(62) # Set the same seed again
print(random.random()) # Will print the same number as before
print(random.random()) # Will print the different number
```

```python
import random

# Set the seed for the random number generator
random.seed(42)  # Using 42 as an example seed value

# Generate a random integer between 1 and 100 (inclusive)
random_number_1 = random.randint(1, 100)
print(f"First random number with seed 42: {random_number_1}")

# Generate another random number
random_number_2 = random.random() # Generates a float between 0.0 and 1.0
print(f"Second random number with seed 42: {random_number_2}")

# If you run the code again with the same seed, you will get the same sequence
print("\nRunning again with the same seed:")
random.seed(42)
random_number_reproducible_1 = random.randint(1, 100)
print(f"First reproducible random number: {random_number_reproducible_1}")

random_number_reproducible_2 = random.random()
print(f"Second reproducible random number: {random_number_reproducible_2}")

# If you don't set a seed, or use a different seed, the numbers will change
print("\nRunning without a specific seed (or with a different one):")
# random.seed() # You can call seed without arguments to use system time as seed
random_number_different_1 = random.randint(1, 100)
print(f"First different random number: {random_number_different_1}")
```
