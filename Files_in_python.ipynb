```python
# Open an existing file in r+ mode
file_path = "example.txt"

# First, make sure the file exists with some content
with open(file_path, 'w') as f:
    f.write("Hello, Python!\nWelcome to file handling.\n")

# Now open it in r+ mode (read + write)
with open(file_path, 'r+') as f:
    # Read the existing content
    print("Original Content:")
    content = f.read()
    print(content)

    # Move the cursor to the end of the file before writing
    f.seek(0, 2)

    # Write additional content
    f.write("Appending new line using r+ mode.\n")

# Verify the changes
with open(file_path, 'r') as f:
    print("\nUpdated File Content:")
    print(f.read())
```

```python
# Create a file for demonstration
with open("my_file.txt", "w") as f:
    f.write("Line 1\nLine 2\nLine 3")

# Open the file in r+ mode
with open("my_file.txt", "r+") as f:
    # Read the entire content
    content = f.read()
    print(f"Initial content:\n{content}")

    # Move the file pointer to the beginning
    f.seek(0)

    # Write new content, overwriting the beginning
    f.write("NEW START")

    # Move the file pointer back to the beginning to read the updated content
    f.seek(0)
    updated_content = f.read()
    print(f"\nContent after writing:\n{updated_content}")

    # Move the file pointer to a specific position and read
    f.seek(5) # Move to the 6th character (index 5)
    partial_read = f.read(4) # Read 4 characters
    print(f"\nPartial read from position 5: {partial_read}")
```

```python
import shutil
import os
# --- Setup: Create a dummy source directory structure ---
source_dir = 'project_source'
destination_dir = 'project_backup'
subdirectory = os.path.join(source_dir, 'assets')
file_path = os.path.join(source_dir, 'config.txt')
nested_file_path = os.path.join(subdirectory, 'image.png')
# Create the directories and dummy files
os.makedirs(subdirectory, exist_ok=True)
with open(file_path, 'w') as f:
    f.write("Configuration settings.")
with open(nested_file_path, 'w') as f:
    f.write("Binary image data placeholder.")
print(f"Source structure created in: {source_dir}")
# --------------------------------------------------------
try:
    # Copy the entire directory tree
    # dirs_exist_ok=False is the default behavior and will raise an error
    # if the destination_dir already exists.
    shutil.copytree(source_dir, destination_dir)

    print(f"\nDirectory tree '{source_dir}' copied successfully to '{destination_dir}'.")

    # Verify contents of the backup directory
    print("\nContents of the backup directory:")
    for root, dirs, files in os.walk(destination_dir):
        level = root.replace(destination_dir, '').count(os.sep)
        indent = ' ' * 4 * level
        print(f'{indent}{os.path.basename(root)}/')
        for f in files:
            print(f'{indent}    {f}')
except FileExistsError:
    print(f"\nError: Destination directory '{destination_dir}' already exists. Cannot copy.")
    print("To allow overwriting partial contents, use shutil.copytree(src, dst, dirs_exist_ok=True) in Python 3.8+.")
except shutil.Error as e:
    print(f"\nAn error occurred during copy operation: {e}")
except Exception as e:
    print(f"\nAn unexpected error occurred: {e}")
# --- Optional Cleanup (uncomment to run) ---
# shutil.rmtree(source_dir)
# shutil.rmtree(destination_dir)
```

```python
import shutil
import os
import time

# --- Setup: Create dummy files and a ZIP archive ---
source_dir = 'content_to_archive'
archive_name = 'my_archive'
archive_format = 'zip'
archive_path = f"{archive_name}.{archive_format}"
extraction_dir = 'extracted_content'

# Create dummy content
os.makedirs(source_dir, exist_ok=True)
with open(os.path.join(source_dir, 'file_a.txt'), 'w') as f:
    f.write("This is file A.")
with open(os.path.join(source_dir, 'file_b.txt'), 'w') as f:
    f.write("This is file B.")

print(f"Created source content in: {source_dir}")

# Create the archive file
print(f"Creating archive '{archive_path}'...")
shutil.make_archive(archive_name, archive_format, root_dir=source_dir)

# Give the filesystem a moment to sync
time.sleep(0.5)
# ----------------------------------------------------

print(f"\nAttempting to unpack '{archive_path}' to '{extraction_dir}'...")

try:
    # Unpack the archive to the destination directory
    # If the destination directory does not exist, it will be created automatically.
    shutil.unpack_archive(archive_path, extraction_dir)

    print("Unpacking successful.")

    # Verify the contents of the extraction directory
    print(f"Contents of '{extraction_dir}': {os.listdir(extraction_dir)}")

except shutil.ReadError:
    print(f"Error: The archive '{archive_path}' is corrupt or an unsupported format.")
except FileNotFoundError:
    print(f"Error: The archive file '{archive_path}' was not found.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")

# --- Optional Cleanup (uncomment to run) ---
# shutil.rmtree(source_dir)
# shutil.rmtree(extraction_dir)
# os.remove(archive_path)
```

```python
import os
import re
import shutil
def rename_american_to_european_dates(directory):
    """
    Renames files in a given directory that have American-style dates (MM-DD-YYYY)
    in their filenames to European-style dates (DD-MM-YYYY).
    Args:
        directory (str): The path to the directory containing the files.
    """
    # Regex to find American-style dates (MM-DD-YYYY) in filenames
    # Group 1: Month (MM), Group 2: Day (DD), Group 3: Year (YYYY)
    date_pattern = re.compile(r"""
        ^(.*?)          # All text before the date (Group 1)
        (\d{1,2})-      # Month (1 or 2 digits) (Group 2)
        (\d{1,2})-      # Day (1 or 2 digits) (Group 3)
        (\d{4})         # Year (4 digits) (Group 4)
        (.*?)$          # All text after the date (Group 5)
        """, re.VERBOSE)
    # Create a new directory
    try:
        os.mkdir("spam04-24-1984.txt")
        print("Directory 'spam4-4-1984.txt' created.")
    except FileExistsError:
        print("Directory 'new_folder' already exists.")
    # Create nested directories
    os.makedirs("parent_folder/child_folder", exist_ok=True)
    print("Nested directories 'parent_folder/child_folder' created.")
    # List contents of a directory
    print(f"Contents of current directory: {os.listdir('.')}")
    for filename in os.listdir(directory):
        mo = date_pattern.search(filename)
        # Skip files without an American-style date in their name
        if not mo:
            continue
        # Get the different parts of the filename
        before_date = mo.group(1)
        month = mo.group(2)
        day = mo.group(3)
        year = mo.group(4)
        after_date = mo.group(5)
        # Form the European-style filename (DD-MM-YYYY)
        new_filename = f"{before_date}{day}-{month}-{year}{after_date}"
        # Get the full, absolute file paths
        abs_working_dir = os.path.abspath(directory)
        amer_filepath = os.path.join(abs_working_dir, filename)
        euro_filepath = os.path.join(abs_working_dir, new_filename)
        # Rename the files
        print(f'Renaming "{amer_filepath}" to "{euro_filepath}"...')
        # Uncomment the following line to actually perform the renaming
        shutil.move(amer_filepath, euro_filepath)
if __name__ == "__main__":
    target_directory = input("Enter the directory containing files to rename: ")
    if os.path.isdir(target_directory):
        rename_american_to_european_dates(target_directory)
        print("Renaming process completed (if 'shutil.move' was uncommented).")
    else:
        print("Invalid directory path provided.")
```

```python
# Define some data
name = "Alice"
age = 30
city = "New York"
temperature = 25.789

# Create a formatted string using an f-string
# You can include variables directly and apply formatting specifiers
formatted_string = f"Name: {name}, Age: {age} years old, City: {city}. Current temperature: {temperature:.2f}C."

# Define the filename
filename = "output.txt"

# Open the file in write mode ('w') and write the formatted string
# The 'with' statement ensures the file is properly closed even if errors occur
try:
    with open(filename, 'w') as file:
        file.write(formatted_string)
    print(f"Successfully wrote the formatted string to '{filename}'.")
except IOError as e:
    print(f"Error writing to file '{filename}': {e}")

# Optionally, read the content back to verify
try:
    with open(filename, 'r') as file:
        content = file.read()
    print("\nContent of the file:")
    print(content)
except IOError as e:
    print(f"Error reading from file '{filename}': {e}")
```

```python
import os

def get_file_info(filepath):
    """
    Calculates the size of a file in bytes and the number of lines it contains.

    Args:
        filepath (str): The path to the file.

    Returns:
        tuple: A tuple containing the file size in bytes and the number of lines.
               Returns (None, None) if the file does not exist or an error occurs.
    """
    if not os.path.exists(filepath):
        print(f"Error: File '{filepath}' not found.")
        return None, None

    try:
        # Get file size in bytes
        file_size_bytes = os.path.getsize(filepath)

        # Count lines
        line_count = 0
        with open(filepath, 'r', encoding='utf-8', errors='ignore') as f:
            for _ in f:
                line_count += 1

        return file_size_bytes, line_count

    except Exception as e:
        print(f"An error occurred: {e}")
        return None, None

# Example usage:
file_to_analyze = "my_document.txt"  # Replace with your file path

# Create a dummy file for demonstration
with open(file_to_analyze, "w") as f:
    f.write("This is line 1.\n")
    f.write("This is line 2.\n")
    f.write("This is line 3.\n")
    f.write("Another line.\n")

size, lines = get_file_info(file_to_analyze)

if size is not None and lines is not None:
    print(f"File: {file_to_analyze}")
    print(f"Size: {size} bytes")
    print(f"Lines: {lines}")

# Clean up the dummy file
os.remove(file_to_analyze)
```

Using csv.DictReader and csv.DictWriter:
These classes are useful when you want to treat each row as a dictionary, with header values as keys.

```python
import csv

filename = "output.csv"

with open(filename, 'r', newline='') as csvfile:
    csv_dict_reader = csv.DictReader(csvfile)
    for row in csv_dict_reader:
        print(row) # Each row is an OrderedDict
        print(row['Name'], row['Age'])
```

```python
import csv

fieldnames = ["Name", "Age", "City"]
data = [
    {"Name": "Charlie", "Age": 35, "City": "Chicago"},
    {"Name": "David", "Age": 28, "City": "Houston"}
]

output_filename = "dict_output.csv"

with open(output_filename, 'w', newline='') as csvfile:
    csv_dict_writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    csv_dict_writer.writeheader() # Writes the header row
    csv_dict_writer.writerows(data)
with open(filename, 'r', newline='') as csvfile:
    csv_dict_reader = csv.DictReader(csvfile)
    for row in csv_dict_reader:
        print(row) # Each row is an OrderedDict
        print(row['Name'], row['Age'])
```

```python
# Function to calculate the area of a rectangle
def calculate_rectangle_area(length, width):
    # Assertion to check that the length and width are positive
    assert length > 0 and width > 0, "Length and width"+ \
                "must be positive"
    # Calculation of the area
    area = length * width
    # Return statement
    return area

# Calling the function with positive inputs
area1 = calculate_rectangle_area(5, 6)
print("Area of rectangle with length 5 and width 6 is", area1)

# Calling the function with negative inputs
area2 = calculate_rectangle_area(-5, 6)
print("Area of rectangle with length -5 and width 6 is", area2)
```
