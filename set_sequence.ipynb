Union (|) Operator

```python
Days1={"Mon","Tue","Wed","Sat"}
Days2={"Thr","Fri","Sat","Sun","Mon"}
print("The Days1 union Days2 is:",Days1 | Days2)
print("The Days1 union Days2 is:",Days1.union(Days2))
```

Intersection (&) Operator

```python
Days1={"Mon","Tue","Wed","Sat"}
Days2={"Thr","Fri","Sat","Sun","Mon"}
print("The Days1 intersection Days2 is:",Days1 & Days2)
print("The Days1 intersection Days2 is:",Days1.intersection(Days2))
```

Difference (-) Operator

```python
Days1={"Mon","Tue","Wed","Sat"}
Days2={"Thr","Fri","Sat","Sun","Mon"}
print("Days1-Days2 is:",Days1 - Days2)
print("Days2-Days1 is:",Days2 - Days1)
print("Days1-Days2 is:",Days1.difference(Days2))
print("Days2-Days1 is:",Days2.difference(Days1))
```

Sorted()

```python
num={10,30,20,40,60,50}
print("The original set num is :",num)
num2 = sorted(num)
print("The sorted list num2 is :",num2)
print("The sorted list num2 is :",sorted(num))
num2 = set(num2)
print("The sorted set num2 is :",num2)
print("The original set num is :",num)
num3 = sorted(num,reverse = True)
print("The sorted list num3 in descending order using sorted is :",num3)
print("The sorted set num3 in descending order using sorted is :",set(num3))
```

```python
lang={'java','c','python','cpp','sql','SQL'}
print("The original set lang is :",lang)
lang2 = sorted(lang)
print("The sorted list lang2 is :",lang2)
print("The sorted list lang2 is :",sorted(lang))
lang2 = set(lang2)
print("The sorted set lang2 is :",lang2)
print("The original set lang is :",lang)
lang3 = sorted(lang,reverse = True)
print("The sorted set lang3 in descending order using sorted is :",set(lang3))
print("The original set lang is :",lang)
```

```python
original_set = {1, 2, 3, 4}
copied_set = original_set.copy()

print(f"Original set: {original_set}")
print(f"Copied set: {copied_set}")

# Demonstrate independence
copied_set.add(5)
print(f"Copied set after modification: {copied_set}")
print(f"Original set after modification of copy: {original_set}")
```

```python
original_set = {'apple', 'banana', 'cherry'}
copied_set = set(original_set)

print(f"Original set: {original_set}")
print(f"Copied set: {copied_set}")

# Demonstrate independence
copied_set.remove('banana')
print(f"Copied set after modification: {copied_set}")
print(f"Original set after modification of copy: {original_set}")
```
