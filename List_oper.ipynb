sorted() sort() append() for numbered list

```python
num=[1,3,2,4,6,5]
print("The original list num is :",num)
num2 = sorted(num)
print("The sorted list num2 is :",num2)
print("The sorted list num2 is :",sorted(num))
print("The original list num is :",num)
print("The sorted list num2 in descending order using sorted is :",sorted(num,reverse = True))
print("The sort list num returning None:",num.sort())
print("The original updated sorted list using sort function is:",num)
num.sort(reverse=True)
print("The updated sorted list (descending order) using sort function is:",num)
num.append(10)
print("The appended list num is:",num)
```

sorted() sort() append() for string list

```python
lang=['java','c','python','cpp','sql','SQL']
lang2 = sorted(lang)
print("The sorted list lang2 is :",lang2)
print("The sorted list lang2 is :",sorted(lang))
print("The original list lang is :",lang)
print("The sort list lang returning None:",lang.sort())
print("The sorted list lang2 in descending order using sorted is :",sorted(lang,reverse = True))
print("The original updated sorted list using sort function is:",lang)
lang.sort(reverse=True)
print("The updated sorted list (descending order) using sort function is:",lang)
fruits=["Banana", "apple", "Cherry"]
fruits2 = sorted(fruits)
print("The sorted list fruits2 is :",fruits2)
print("The sorted list fruits3 is :",sorted(fruits, key=str.lower))
print("The sorted list based on length fruits4 is :",sorted(fruits, key=len))
print("The sorted list based on last char fruits4 is :",sorted(fruits, key=lambda s:s[-1]))
print("The sorted list based on last before char fruits4 is :",sorted(fruits, key=lambda s:s[-2]))
print("The sorted list based on char[1] fruits4 is :",sorted(fruits, key=lambda s:s[1]))
lang.append("perl")
print(lang)
```

sorting list with numbers and strings

```python
list1 = [1, 2, 3, 4, 5, 'python', 'c', 'java', 'php']

# Separate integers and strings
num_list = [item for item in list1 if isinstance(item, int)]
lang_list = [item for item in list1 if isinstance(item, str)]

# Sort the lists separately
sorted_num = sorted(num_list)
sorted_lang = sorted(lang_list)

print("Original list:", list1)
print("Sorted numbers:", sorted_num)
print("Sorted strings:", sorted_lang)
```

reverse() or flipping elements in list

```python
lang = ['p', 'y', 't', 'h', 'o','n'] # Char list
even = [6,8,2,4,10] # int list
print("The lang list is:",lang)
print("The even num list is:",even)
lang.reverse()
print("The reversed lang list is:",lang)
even.reverse()
print("The reversed even num list is:",even)
```

```python
nums = [1, 2, 3, 4, 5]

nums.insert(2, 'Hello')  # Insert at index position 1.
# Result: [1, 'Hello', 2, 3, 4, 5]
print(nums)

nums.insert(-1, 'Bye')  # Insert at index position len(nums)-1 = 5.
# Same as: nums.insert(5, 'Bye')
# Result: [1, 'Hello', 2, 3, 4, 'Bye', 5]
print(nums)
nums.insert(-2, 'Good')  # Insert at index position len(nums)-2 = 5.
# Same as: nums.insert(5, 'Good')
# Result: [1, 'Hello', 2, 3, 4, 'Good', 'Bye', 5]
print(nums)
```

Magic 8 ball code

```python
import random
messages = ['It is certain',
'It is decidedly so',
'Yes definitely',
'Reply hazy try again',
'Ask again later',
'Concentrate and ask again',
'My reply is no',
'Outlook not so good',
'Very doubtful']
print(messages[random.randint(0, len(messages) - 1)])
```

```python
# code
while True:
       question = input("Ask the magical 8 Ball a question (type 'exit' to quit): ")
       if question.lower() == 'exit':
           print("Goodbye!")
           break
       else:
           print("Magic 8 Ball says:", random.choice(messages))
```
