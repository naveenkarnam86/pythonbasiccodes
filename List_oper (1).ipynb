sorted() sort() append() for numbered list

```python
num=[1,3,2,4,6,5]
print("The original list num is :",num)
num2 = sorted(num)
print("The sorted list num2 is :",num2)
print("The sorted list num2 is :",sorted(num))
print("The original list num is :",num)
print("The sorted list num2 in descending order using sorted is :",sorted(num,reverse = True))
print("The sort list num returning None:",num.sort())
print("The original updated sorted list using sort function is:",num)
num.sort(reverse=True)
print("The updated sorted list (descending order) using sort function is:",num)
num.append(10)
print("The appended list num is:",num)
```

sorting list with numbers and strings

```python
list1 = [1, 2, 3, 4, 5, 'python', 'c', 'java', 'php']

# Separate integers and strings
num_list = [item for item in list1 if isinstance(item, int)]
lang_list = [item for item in list1 if isinstance(item, str)]

# Sort the lists separately
sorted_num = sorted(num_list)
sorted_lang = sorted(lang_list)

print("Original list:", list1)
print("Sorted numbers:", sorted_num)
print("Sorted strings:", sorted_lang)
```

list shallow copy

```python
original_list = [1, 2, [3, 4]]
copied_list = original_list.copy()
print("Original List:", original_list)
print("Copied List:", copied_list)
# modifying an element in copy
copied_list[1] = 7
print("Original List after modification:", original_list)
print("Copied List after modification:", copied_list)
# modifying a nested element in copy
copied_list[2][0] = 5
print("Original List after modification:", original_list)
print("Copied List after modification:", copied_list)
# appending a nested element in copy
copied_list.append(6)
print("Original List after modification:", original_list)
print("Copied List after modification:", copied_list)
```

```python
a = [1, 'two', 3.0, [4, 5], {'key': 'value'}]
b = a.copy()

b[3].append(6)

print("a:", a)
print("b:", b)
```

Using list slicing: Slicing the entire list creates a new list containing references to the original elements.

```python
original_list = [1, 2, [3, 4]]
copied_list = original_list[:]
print("Original List:", original_list)
print("Copied List:", copied_list)
```

Using the list() constructor: Passing the original list to the list() constructor also creates a shallow copy.

```python
original_list = [1, 2, [3, 4]]
copied_list = list(original_list)
print("Original List:", original_list)
print("Copied List:", copied_list)
```

Using the list comprehension

```python
original_list = [1, 2, [3, 4]]
copied_list = [item for item in original_list]
print("Original List:", original_list)
print("Copied List:", copied_list)
```

deepcopy

```python
import copy

a = [[1, 2], [3, 4], [5, 6]]
print("Original List:",a)
# Copying the list using deepcopy
b = copy.deepcopy(a)
print("copied List:",b)
# modifying a nested element in copy
b[0][0] = 10
print("copied List:",b)
print("Original List:",a)
# appending an element in copy
b.append(7)
print("copied List",b)
print("Original List:",a)
```

Magic 8 ball code

```python
import random
messages = ['It is certain',
'It is decidedly so',
'Yes definitely',
'Reply hazy try again',
'Ask again later',
'Concentrate and ask again',
'My reply is no',
'Outlook not so good',
'Very doubtful']
print(messages[random.randint(0, len(messages) - 1)])
```

```python
# code
while True:
       question = input("Ask the magical 8 Ball a question (type 'exit' to quit): ")
       if question.lower() == 'exit':
           print("Goodbye!")
           break
       else:
           print("Magic 8 Ball says:", random.choice(messages))
```
